```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36',
           'Accept-Language': 'en-US',
           'Accept-Encoding': 'gzip, deflate',
           'Accept': 'text/html',
           'Referer': 'http://www.google.com/'}

url = 'https://www.coupang.com/np/categories/194282?page={}'

fruits = []

for page in range(1, 16):
    response = requests.get(url.format(page), headers=headers)
    if response.status_code == 200:
        fruits_list = bs(response.text).find_all('li', {'class': 'baby-product renew-badge'})
        for fruit in fruits_list:
            link = fruit.find('a', {'class': 'baby-product-link'})['href'].split('?')[0]
            name = fruit.find('div', {'class': 'name'}).text.strip()
            price = fruit.find('strong', {'class': 'price-value'}).text
            rating = fruit.find('em', {'class': 'rating'}).text
            rating_total_count = fruit.find('span', {'class': 'rating-total-count'}).text
            fruits.append((link, name, price, rating, rating_total_count))
    else:
        print('error')
        
df = pd.DataFrame(fruits, columns=['link', 'name', 'price', 'rating', 'rating_total_count'])
df.head()
```
